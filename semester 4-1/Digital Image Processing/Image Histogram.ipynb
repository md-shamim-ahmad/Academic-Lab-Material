# Image Histogram Without using imhist or hist Library

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt 
```

```python
img = Image.open('image2.jpg')
img = img.convert('L')
img_arr = np.array(img)
img_arr.shape
```

```python
histogram = {}
for pixel in img_arr:
    for value in pixel:
        if value in histogram:
            histogram[value] += 1
        else:
            histogram[value] = 1

histogram = dict(sorted(histogram.items()))
```

```python
histo_img = np.zeros(256, int)
for key, value in histogram.items():
    histo_img[key] = value    
```

```python
# Plot the histogram
plt.figure(figsize=(10, 5))
plt.bar(np.arange(len(histo_img)), histo_img)
plt.title('Image Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```
