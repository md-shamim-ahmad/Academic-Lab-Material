```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image1 = Image.open('image3.jpg').convert('L')
image2 = Image.open('image2.jpg').convert('L')

image1_array = np.array(image1)
image2_array = np.array(image2)
```

# Image Addition

```python
if image1_array.shape != image2_array.shape:
    image2_array = np.array(image2.resize(image1.size))

result_array = np.clip(image1_array + image2_array, 0, 255).astype(np.uint8)
result_image = Image.fromarray(result_array)

# Display the original images and the result
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image1_array, cmap='gray')
plt.title('Image 1')
# plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image2_array, cmap='gray')
plt.title('Image 2')

plt.subplot(1, 3, 3)
plt.imshow(result_image, cmap='gray')
plt.title('Image Addition Result')

plt.show()
```

# Image Subtraction

```python
if image1_array.shape != image2_array.shape:
    image2_array = np.array(image2.resize(image1.size))

result_array = np.clip(image1_array - image2_array, 0, 255).astype(np.uint8)
result_image = Image.fromarray(result_array)

# Display the original images and the result
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image1_array, cmap='gray')
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image2_array, cmap='gray')
plt.title('Image 2')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(result_image, cmap='gray')
plt.title('Image Subtraction Result')
plt.axis('off')

plt.show()
```

# Image Multiplaction

```python
if image1_array.shape != image2_array.shape:
    image2_array = np.array(image2.resize(image1.size))

result_array = np.clip(image1_array * image2_array, 0, 255).astype(np.uint8)
result_image = Image.fromarray(result_array)

# Display the original images and the result
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image1_array, cmap='gray')
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image2_array, cmap='gray')
plt.title('Image 2')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(result_image, cmap='gray')
plt.title('Image Multiplaction Result')
plt.axis('off')

plt.show()
```

# Image Division

```python
if image1_array.shape != image2_array.shape:
    image2_array = np.array(image2.resize(image1.size))

epsilon = 1e-5
result_array = np.clip(image1_array / (image2_array + epsilon), 0, 255).astype(np.uint8)
result_image = Image.fromarray(result_array)


# Display the original images and the result

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image1_array, cmap='gray')
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image2_array, cmap='gray')
plt.title('Image 2')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(result_image, cmap='gray')
plt.title('Image Division Result')
plt.axis('off')

plt.show()
```
